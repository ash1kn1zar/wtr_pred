```python
X= weather[['min_temp', 'max_temp', 'precipitation', "max_wind_speed", "most_wind_direction", "avg_relative_humidity"]]
```

```python
X.head()
```

```python
y=weather[['avg_temp']]
```

```python
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
fitted = scaler.fit(X)
transformed = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(transformed, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
predictions = lr.predict(X_test)
print(predictions)
```

```python
predicted_values = pd.DataFrame({'Predicted_values': predictions[:, 0]})
predicted_values.describe()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
score = lr.score(X_test, y_test)
print(score)
```
